# Cleaning texts

```python
import re
with open ('tex1.txt', 'r', encoding="utf8") as f:
    content = f.read()
    content_1 = re.sub('\-\d*\-|\- \d* \-|\n',' ',content)
    
import re
with open ('tex2.txt', 'r', encoding="utf8") as f:
    content1 = f.read()
    content_2 = re.sub('\-\d*\-|\- \d* \-|\n',' ',content1)
```

```python
content_1
```

# To understand the differences, splitted from fullstop each line

```python
z1=re.split('[.]',content_1)
z2=re.split('[.]',content_2)
```

## Converting to Pandas dataframes

```python
import pandas as pd
df=pd.DataFrame(z1)
df1=pd.DataFrame(z2)
```
